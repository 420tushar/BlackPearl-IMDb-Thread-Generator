```
import urllib.request
import json
import re
import requests
import sys
import subprocess
import os
import time
import pathlib
import urllib.request
from bs4 import BeautifulSoup

import os, re, time, pathlib
import urllib.request
from IPython.display import clear_output

Auto_UP_Gdrive = False  
AUTO_MOVE_PATH = "/content" 
HOME = os.path.expanduser("~")
pathDoneCMD = f'{HOME}/doneCMD.sh'

if not os.path.exists(f"{HOME}/.ipython/ttmg.py"):
    hCode = "https://raw.githubusercontent.com/yunooooo/gcct/master/res/ttmg.py"
    urllib.request.urlretrieve(hCode, f"{HOME}/.ipython/ttmg.py")

from ttmg import (
    loadingAn,
)

%cd "/content"
%rm -rf "/content/frames"
#@markdown <h6><center><img src="https://i.imgur.com/FaxnTtv.png" height="86"/></center>
#@markdown <center><font size="7" face="Grand Hotel" color="#56b84c">⸻⸻ Posting Template ⸻⸻</center>
#@markdown <center></font></center>

# Variables
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.8; rv:21.0) Gecko/20100101 Firefox/21.0'}
#@markdown <h1><font size="6" face="Montserrat" color="#9FD4F3">OMDb </font><font size="6" face="Nunito">API Key:</font></center>
api_key = '28ab9ae6' #@param {type:"string"}
#@markdown <center>឵឵ ឵</center>

#@markdown <h1><font size="6" face="Montserrat" color="#F1F290">IMDb</font><font size="6" face="Nunito"> URL <small>or</small> ID:</font></center>
IMDB_URL_OR_ID = 'https://www.imdb.com/title/tt0266543' #@param {type:"string"}
#@markdown <center>឵឵ ឵</center>
#@markdown <h1><font size="6" face="Montserrat" color="#F6675E">YouTube</font><font size="6" face="Nunito"> URL:</center>
YouTube_URL = 'https://www.youtube.com/watch?v=9oQ628Seb9w' #@param {type:"string"}
if YouTube_URL == "":
  youtube_enabled = False
else:
  youtube_enabled = True
#@markdown <center>឵឵ ឵</center>
#@markdown <center><font size="7" face="Arsenal" color="#E9A3C4">Movie </font><font face="Oswald" color="#9FA5F3" font size="5">឵឵OR឵</font><font size="7" face="Arsenal" color="#E9A3C4"> TV File Info:</font></div>  
#@markdown <center>឵឵ ឵</center>
Path = "/content/drive/Shared drives/Movies/Finding.Nemo.2003.2160p.BluRay.REMUX.HEVC.TrueHD.7.1.Atmos-FGT.mkv" #@param {type:"string"}
if Path == "":
  mediainfo_enabled = False
else:
  mediainfo_enabled = True
  tmp = subprocess.Popen('mediainfo --Logfile=/root/.nfo "{}"'.format(Path), stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
  time.sleep(5) # Wait 5s for the nfo file to populate properly
  with open('/root/.nfo', 'r', encoding='utf-8') as nfo_file:
    nfo = nfo_file.readlines()
    del nfo[1]
    nfo[1] = "Complete name                            : {}\n".format(os.path.basename(Path))
    nfo = ''.join(nfo)

!mkdir -p "/content/frames"
for i in range(10):
    clear_output()
    loadingAn()

import os, sys, re

output_file_path = re.search("^[\/].+\/", Path)
output_file_path_raw = output_file_path.group(0)
delsplit = re.search("\/(?:.(?!\/))+$", Path)
filename = re.sub("^[\/]", "", delsplit.group(0))
filename_raw = re.sub(".{4}$", "", filename)
file_extension = re.search(".{3}$", filename)
file_extension_raw = file_extension.group(0)

%cd "$output_file_path_raw"
!ffmpeg -hide_banner -ss 01:30.0 -i "$filename" -vframes 1 -q:v 2 -y "/content/frames/frame1.png"
!ffmpeg -hide_banner -ss 03:12.0 -i "$filename" -vframes 1 -q:v 2 -y "/content/frames/frame2.png"
!ffmpeg -hide_banner -ss 05:30.0 -i "$filename" -vframes 1 -q:v 2 -y "/content/frames/frame3.png"
clear_output()

loadingAn()
%cd "/content/frames"
!curl -F "reqtype=fileupload" -F "fileToUpload=@frame1.png" https://catbox.moe/user/api.php -o "/content/frames/frame1.txt"
if os.path.exists("/content/frames/frame1.txt"):
  f1 = open('/content/frames/frame1.txt', 'r')
  file_content1 = f1.read()
  f1.close()
else:
  print("No Frames Detected!")

!curl -F "reqtype=fileupload" -F "fileToUpload=@frame2.png" https://catbox.moe/user/api.php -o "/content/frames/frame2.txt"
if os.path.exists("/content/frames/frame2.txt"):
  f2 = open('/content/frames/frame2.txt', 'r')
  file_content2 = f2.read()
  f2.close()

if os.path.exists("/content/frames/frame3.png"):
  os.system("curl -F 'reqtype=fileupload' -F 'fileToUpload=@frame3.png' https://catbox.moe/user/api.php -o '/content/frames/frame3.txt'")
  f3 = open('/content/frames/frame3.txt', 'r')
  file_content3 = f3.read()
  f3.close()
  framenum3_enabled = True
else:
  framenum3_enabled = False

imdb_pattern = re.compile('(tt\d{7,8})')
imdb_id = re.findall(imdb_pattern, IMDB_URL_OR_ID)[0]

%cd "/content"
def get_imdb_template(imdb_id):
  imdb_url = f'http://www.omdbapi.com/?i={imdb_id}&apikey={api_key}&r=json&plot=full'
  Screenshot1 = file_content1
  Screenshot2 = file_content2
  YouTube_ID = re.search("([A-Za-z0-9-_]){11}", YouTube_URL)

  Link = "https://drive.google.com/open?id=1nvylnho7Nz0eAyebPztyjKsW417Zw8DM" #@param {type:"string"}
  if Link == "":
    print("You forgot putting in the link to download!")
    sys.exit()

  def get_story_line(imdb_id):
    url = 'https://www.imdb.com/title/{}'.format(imdb_id)
    r = requests.get(url, headers=headers)
    data = r.text
    soup = BeautifulSoup(data, 'html.parser')
    story_line = soup.find_all('div', class_='inline canwrap')[0].find('span').text.lstrip()
    return story_line

  # Fetch data
  with urllib.request.urlopen(imdb_url) as imdb_url:
    imdb_data = json.loads(imdb_url.read().decode())

  LQ_poster = "{}".format(imdb_data['Poster'])
  HQ_poster = re.sub("[._V1_SX300]{10}", "", LQ_poster)

  # Process data
  text_dump = "[CENTER][SIZE=7]{} ({})[/SIZE]\n".format(imdb_data['Title'], imdb_data['Year'])
  text_dump += "\n"
  text_dump += "[IMG width='450px']{}[/IMG][/CENTER]\n".format(HQ_poster)
  text_dump += "\n"
  text_dump += "[FONT=Montserrat][SIZE=6][COLOR=#F1F290]IMDb[/COLOR][/SIZE][/FONT]\n"
  text_dump += "\n[INDENT][SIZE=4]Plot: {}[/INDENT]\n".format(imdb_data['Plot'])
  text_dump += "[INDENT]Genre: {}\n".format(imdb_data['Genre'])
  text_dump += "URL: https://www.imdb.com/title/{}[/SIZE][/INDENT]".format(imdb_id)
  text_dump += "\n"
  if youtube_enabled:
    text_dump += "[FONT=Montserrat][COLOR=#E7797E][SIZE=6]Trailer[/size][/COLOR][/FONT]\n"
    text_dump += "\n"
    text_dump += "[INDENT][MEDIA=youtube]{}[/MEDIA][/INDENT]\n".format(YouTube_ID.group(0))
  if mediainfo_enabled:
    text_dump += "\n"
    text_dump += "\n[COLOR=#83A8CE][SIZE=5][I]MediaInfo:[/I][/SIZE][/COLOR]\n"
    text_dump += "[SPOILER=MediaInfo][ICODE]"
    text_dump += "{}".format(nfo)
    text_dump += "[/ICODE][/SPOILER]"
  text_dump += "\n"
  text_dump += "[COLOR=#D5A6D5][SIZE=5][I]Screenshots:[/I][/SIZE][/COLOR]\n"
  text_dump += "[SPOILER=Screenshots][CENTER]\n"
  text_dump += "[IMG width='800px']{}[/IMG]\n".format(Screenshot1)
  text_dump += "\n"
  text_dump += "[IMG width='800px']{}[/IMG]".format(Screenshot2)
  if framenum3_enabled:
    text_dump += "\n"
    text_dump += "\n[IMG width='800px']{}[/IMG][/CENTER]\n".format(file_content3)
  else:
    text_dump += "[/CENTER]\n"
  text_dump += "[/SPOILER]\n"
  text_dump += "[CENTER][COLOR=#39B73C][SIZE=6][FONT=Trebuchet MS]Download Link:[/FONT][/SIZE][/COLOR]\n"
  text_dump += "[HIDEREACT=1][URL={}]{}[/URL][/HIDEREACT][/CENTER]".format(Link, Link)

  return text_dump

clear_output()
print(get_imdb_template(imdb_id))
```
